#### 5. Find below metrics for numerical columns
Number of values, number of missing's, number of distinct values, number of zeros, sum, mean, median, std, variance, min, max, range, P1, P5, P10, P25, P50, P75, P90, P95, P99 
(Note: You should create user defined function for the same)

```python
# Define a function to compute the required metrics
def compute_metrics(df):
    metrics = {}
    numerical_cols = df.select_dtypes(include=[float, int]).columns
    
    for col in numerical_cols:
        col_data = df[col]
        col_metrics = {
            'number_of_values': col_data.count(),
            'number_of_missing': col_data.isnull().sum(),
            'number_of_distinct_values': col_data.nunique(),
            'number_of_zeros': (col_data == 0).sum(),
            'sum': col_data.sum(),
            'mean': col_data.mean(),
            'median': col_data.median(),
            'std': col_data.std(),
            'variance': col_data.var(),
            'min': col_data.min(),
            'max': col_data.max(),
            'range': col_data.max() - col_data.min(),
            'P1': col_data.quantile(0.01),
            'P5': col_data.quantile(0.05),
            'P10': col_data.quantile(0.10),
            'P25': col_data.quantile(0.25),
            'P50': col_data.quantile(0.50),
            'P75': col_data.quantile(0.75),
            'P90': col_data.quantile(0.90),
            'P95': col_data.quantile(0.95),
            'P99': col_data.quantile(0.99)
        }
        metrics[col] = col_metrics
    
    return pd.DataFrame(metrics)

# Compute metrics for numerical columns in the final merged dataframe
metrics_df = compute_metrics(final_data)

# Display the computed metrics
print(metrics_df)
```

####  b. Outlier treatment

```python
final_data = pd.read_csv(r"C:\Users\Bahroz Kalam\Desktop\Python\PYTHON CASE STUDY\final_data.csv")
final_data
def treat_outliers(df):
    numerical_cols = df.select_dtypes(include=[float, int]).columns
    for col in numerical_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
        df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
    
    return df
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency, ttest_ind, f_oneway

# Load the cleaned dataframe
cleaned_data = pd.read_csv(r"C:\Users\Bahroz Kalam\Desktop\Python\PYTHON CASE STUDY\final_data.csv")
cleaned_data

# Check the first few rows of the dataset
print(cleaned_data.head())

# Define a function for Chi-Square test for categorical variables
def chi_square_test(df, target_var):
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns
    results = {}
    for col in categorical_cols:
        contingency_table = pd.crosstab(df[col], df[target_var])
        chi2, p, dof, ex = chi2_contingency(contingency_table)
        results[col] = p
    return results

# Define a function for T-Test or ANOVA for numerical variables
def t_test_anova(df, target_var):
    numerical_cols = df.select_dtypes(include=[float, int, bool]).columns
    results = {}
    for col in numerical_cols:
        group1 = df[df[target_var] == 0][col]
        group2 = df[df[target_var] == 1][col]
        if df[target_var].nunique() == 2:
            t_stat, p_val = ttest_ind(group1.dropna(), group2.dropna())
        else:
            t_stat, p_val = f_oneway(group1.dropna(), group2.dropna())
        results[col] = p_val
    return results

# Perform Chi-Square test for categorical variables
chi_square_results = chi_square_test(cleaned_data, 'attrition')

# Perform T-Test or ANOVA for numerical variables
t_test_anova_results = t_test_anova(cleaned_data, 'attrition')

# Combine and display the results
print("Chi-Square Test Results for Categorical Variables:")
for col, p_val in chi_square_results.items():
    print(f"{col}: p-value = {p_val}")

print("\nT-Test/ANOVA Results for Numerical Variables:")
for col, p_val in t_test_anova_results.items():
    print(f"{col}: p-value = {p_val}")

# Plotting significant variables
significant_categorical_vars = [col for col, p_val in chi_square_results.items() if p_val < 0.05]
significant_numerical_vars = [col for col, p_val in t_test_anova_results.items() if p_val < 0.05]

# Plot bar charts for significant categorical variables
for col in significant_categorical_vars:
    plt.figure(figsize=(12, 6))
    sns.countplot(x=col, hue='attrition', data=cleaned_data, palette='viridis')
    plt.title(f'{col}')
    plt.show()

# Plot box plots for significant numerical variables
for col in significant_numerical_vars:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='attrition', y=col, data=cleaned_data, palette='viridis')
    plt.title(f'{col}')
    plt.show()
```

#### c. How employee job related variables influencing attrition? - 10 Marks


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency, ttest_ind, f_oneway

# Load the cleaned dataframe
cleaned_data = pd.read_csv(r"C:\Users\Bahroz Kalam\Desktop\Python\PYTHON CASE STUDY\final_data.csv")
cleaned_data

# Identify job-related variables
job_related_vars = [
    'joblevel', 'jobrole', 'department', 'monthlysalary', 
    'yearsatcompany', 'yearsincurrentrole', 'yearswithcurrmanager'
]

# Check the first few rows of the dataset
print(cleaned_data.head())

# Define a function for Chi-Square test for categorical variables
def chi_square_test(df, target_var, job_related_vars):
    results = {}
    for col in job_related_vars:
        if df[col].dtype == 'object' or df[col].dtype.name == 'category':
            contingency_table = pd.crosstab(df[col], df[target_var])
            chi2, p, dof, ex = chi2_contingency(contingency_table)
            results[col] = p
    return results

# Define a function for T-Test or ANOVA for numerical variables
def t_test_anova(df, target_var, job_related_vars):
    results = {}
    for col in job_related_vars:
        if df[col].dtype in [float, int]:
            group1 = df[df[target_var] == 0][col]
            group2 = df[df[target_var] == 1][col]
            if df[target_var].nunique() == 2:
                t_stat, p_val = ttest_ind(group1.dropna(), group2.dropna())
            else:
                t_stat, p_val = f_oneway(group1.dropna(), group2.dropna())
            results[col] = p_val
    return results

# Perform Chi-Square test for categorical job-related variables
chi_square_results = chi_square_test(cleaned_data, 'attrition', job_related_vars)

# Perform T-Test or ANOVA for numerical job-related variables
t_test_anova_results = t_test_anova(cleaned_data, 'attrition', job_related_vars)

# Combine and display the results
print("Chi-Square Test Results for Categorical Job-Related Variables:")
for col, p_val in chi_square_results.items():
    print(f"{col}: p-value = {p_val}")

print("\nT-Test/ANOVA Results for Numerical Job-Related Variables:")
for col, p_val in t_test_anova_results.items():
    print(f"{col}: p-value = {p_val}")

# Plotting significant job-related variables
significant_categorical_vars = [col for col, p_val in chi_square_results.items() if p_val < 0.05]
significant_numerical_vars = [col for col, p_val in t_test_anova_results.items() if p_val < 0.05]

# Plot bar charts for significant categorical variables
for col in significant_categorical_vars:
    plt.figure(figsize=(12, 6))
    sns.countplot(x=col, hue='attrition', data=cleaned_data, palette='viridis')
    plt.title(f'{col}')
    plt.show()

# Plot box plots for significant numerical variables
for col in significant_numerical_vars:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='attrition', y=col, data=cleaned_data, palette='viridis')
    plt.title(f'{col}')
    plt.show()
```
